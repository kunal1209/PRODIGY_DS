```python
#import libraries
import pandas as pd
from pandas_profiling import ProfileReport
```

**DATA CLEANING**

```python
#LOAD THE DATASET
df = pd.read_excel('/content/uncleaned bike sales data.xlsx', sheet_name='Bike Sales')
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
#HANDLING MISSING VALUES:
 # Fill with median or any other suitable value
df['Day'].fillna(df['Day'].median(), inplace=True)

# Fill with the most frequent category
df['Age_Group'].fillna(df['Age_Group'].mode()[0], inplace=True)
df['Product_Description'].fillna('Unknown', inplace=True)

 # Fill with a suitable placeholder
df['Order_Quantity'].fillna(df['Order_Quantity'].median(), inplace=True)

df.isnull().sum()
```

```python
#checking for inconsistent data
df.dtypes
```

```python
#Handling inconsistent data

#convert into int as these are typically whole number
df['Day'] = df['Day'].astype(int)

df['Order_Quantity'] = df['Order_Quantity'].astype(int)

```

```python
#checking for duplicates
df.duplicated()
df.to_csv('cleaned_dataset.csv')
```

***Exploratory Data Analysis using Pandas Profiling***

```python
#LOAD CLEAN DATA TO PERFORM EDA
df = pd.read_csv('/content/cleaned_dataset.csv')

# Generate a Pandas Profiling report
profile = ProfileReport(df, title='DATA ANALYSIS REPORT', explorative=True)

profile.to_file('eda_report.html')
```
